```python
plt.plot(df_yahoo['High'], label='High')
plt.plot(df_yahoo['Low'], label='Low')
plt.legend()
plt.show()
```

```python
df_yahoo['Media'] = df_yahoo['High'] + df_yahoo['Low']
df_yahoo['Media'] = df_yahoo['Media'] / 2
df_yahoo['Media']
```

```python
decomposicao = seasonal_decompose(df_yahoo['Media'], model='additive', period=12)
```

```python
trend = decomposicao.trend
```

```python
seasonal = decomposicao.seasonal
```

```python
aleatorio = decomposicao.resid
```

```python
plt.figure(figsize=(12, 10))

# Original
plt.subplot(4,1,1)
plt.plot(decomposicao.observed, label="Original", color="blue")
plt.legend()

# Tendncia
plt.subplot(4,1,2)
plt.plot(decomposicao.trend, label="Tendncia", color="orange")
plt.legend()

# Sazonalidade
plt.subplot(4,1,3)
plt.plot(decomposicao.seasonal, label="Sazonalidade", color="green")
plt.legend()

# Resduo
plt.subplot(4,1,4)
plt.plot(decomposicao.resid, label="Aleatoriedade", color="red")
plt.legend()

plt.tight_layout()  # ajusta os espaos entre os subplots
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(df_yahoo['Media'], order=(1,1,1), seasonal_order=(1,1,1,12))
result = model.fit()
```

```python
print(result.summary())
```

```python
predict = result.get_forecast(steps=12)
mean_predict = predict.predicted_mean
interval_predict = predict.conf_int()
```

```python
df_yahoo.index.values
```

```python
plt.figure(figsize=(12,6))

plt.plot(df_yahoo.index, df_yahoo['Media'], label="Historico")
plt.plot(mean_predict.index, mean_predict, label = "Predict", color="red")
plt.plot(interval_predict.index, interval_predict.iloc[:,0],
         interval_predict.iloc[:,1], color="pink", alpha=0.3)
plt.legend()
plt.show()
```

```python
filter_interval = interval_predict[interval_predict.index >= "2020-01-01"]
```

```python
plt.figure(figsize=(12,6))

plt.plot(df_yahoo.index, df_yahoo['Media'], label="Historico")
plt.plot(mean_predict.index, mean_predict, label = "Predict", color="red")
plt.plot(filter_interval.index, filter_interval.iloc[:,0],
         filter_interval.iloc[:,1], color="pink", alpha=0.3)
plt.legend()
plt.show()
```
